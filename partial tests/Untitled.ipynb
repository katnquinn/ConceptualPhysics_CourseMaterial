```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,12))
plt.plot([0,1,40,41,52,56,60],[0,4,4,0,0,-40,0],linewidth=2,color='red')
plt.plot([0,60],[0,0],linewidth=2,color='black')
plt.xticks([i for i in range(61)])
plt.yticks([i for i in range(-40,11)])
plt.xlim([0,60])
plt.ylim([-40,10])
plt.ylabel('Velocity (m/s)',fontsize=16)
plt.xlabel('Time (s)',fontsize=16)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,10))

theta = np.linspace(0,4*np.pi,10000)
neutron = np.sin(theta*12)
proton = np.sin(theta*9)+3
electron = np.sin(theta*2) + 6
antineutrino = np.sin(theta) + 9


plt.plot(theta,neutron)
plt.plot(theta,proton)
plt.plot(theta,electron)
plt.plot(theta,antineutrino)

plt.xticks([])
plt.yticks([0,3,6,9],['(d) neutron','(c) proton','(b) electron','(a) antineutrino'])


plt.show()
```
